```python
#創世紀
from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup

output = ""
print("創世紀")
i = 1
while (i<=51):
  url = "https://cb.fhl.net/read.php?chineses=創&chap="+str(i)+"&VERSION6=sed&TABFLAG=1"
  res = requests.get(url)
  res.encoding = "utf-8"
  html = res.text
  soup = BeautifulSoup(html, "html.parser")
  # print(soup.prettify())
  context = soup.find_all("td")
  for context in context:
    print(context.text)
  i = i+1
```
